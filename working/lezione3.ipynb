# Q2: Planetary positions

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06, 39.48]) #distanza
P = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.09]) #periodo 
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", 
         "Uranus", "Neptune", "Pluto"] #nome



fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(a, P,'o', label="period")

ax.set_xlabel("Distances",fontsize="large")
ax.set_ylabel("Period", fontsize="large")
ax.set_xscale('log')
ax.set_yscale('log')

for i in range(len(a)): 
    ax.text(a[i], P[i], names[i])

plt.show()
```

# Q3: A perfect circle

```python
import numpy as np
import matplotlib.pyplot as plt
import random

R = 5
theta = np.linspace(0, 2 * np.pi, 100)
x = []
y = []
for i in range(len(theta)):
    x.append(R * np.cos(theta[i]))
    y.append(R * np.sin(theta[i]))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y,'o', label="period")

ax.set_xlabel("Distance",fontsize="large")
ax.set_ylabel("Period", fontsize="large")

plt.fill(x, y, color='red', alpha=0.2)

ax.set_aspect('equal')



def draw_circle(x0, y0, R, color):
   
    theta = np.linspace(0, 2 * np.pi, 500)
    x = x0 + R * np.cos(theta)
    y = y0 + R * np.sin(theta)
    plt.fill(x, y, color=color, alpha=0.4) 

plt.figure(figsize=(8, 8))


for _ in range(10):
    x0 = random.uniform(-20, 20)  # Centro x 
    y0 = random.uniform(-20, 20)  # Centro y 
    R1 = random.uniform(1, 10)      # Raggio
    color = np.random.rand(3,)    # Colore RGB casuale
    draw_circle(x0, y0, R1, color)

# Impostazioni del grafico

plt.title("10 Cerchi Casuali")
plt.xlabel("Distance")
plt.ylabel("Period")
plt.show()
```

# Q4: Climate

```python
import numpy as np
import matplotlib.pyplot as plt
import requests


url = "https://raw.githubusercontent.com/sbu-python-summer/python-tutorial/master/day-4/nasa-giss.txt"
filename = "nasa-giss.txt"

response = requests.get(url)
with open(filename, 'wb') as f:
    f.write(response.content)


x,y,z = np.loadtxt(filename,unpack = True)


fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, z, label="T")

y_pos = np.where(y > 0, y, np.nan )
y_neg = np.where(y < 0, y, np.nan)

ax.plot(x, y_pos,'o', color="red", label="smoothed  T > 0" )
ax.plot(x, y_neg,'o', color="blue", label="smoothed  T < 0" )

ax.set_xlabel("Years",fontsize="large")
ax.set_ylabel("T", fontsize="large")


ax.legend()
plt.show()
```
